```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

df = pd.read_csv("../data/cleaned_ckd_data.csv")
df.head()
```

## Distribution of Age


```python
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')
plt.title("Distribution of Age")
plt.show()
```

## Hypertension Count


```python
sns.countplot(
    x='FamilyHistoryHypertension',
    hue='FamilyHistoryHypertension',
    data=df,
    palette='pastel',
    legend=False
)
plt.title("Family History of Hypertension")
plt.xlabel("Family History of Hypertension")
plt.ylabel("Count")
plt.show()
```

## Correlation Heatmap


```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## BMI by Diagnosis


```python
sns.boxplot(
    x='Diagnosis',
    y='BMI',
    hue='Diagnosis',
    data=df,
    palette='Set2',
    legend=False
)
plt.title("BMI by Diagnosis")
plt.show()
```
